```python
import geopandas as gpd
import Functions
from pathlib import Path
import pandas as pd

import importlib

importlib.reload(Functions)
```

```python
app_path = Functions.get_input_path() / 'App'

input_path = app_path / 'Data' / 'Ground_Campaign'
shape_path = input_path / 'Flevoland_data' / 'Data_25_fields' / 'Flevoland-fields-Shapefiles' / 'Feloveland-fields-Shapefiles'
shapes = list(shape_path.glob('*.shp'))
```

```python
data_list = []


for file in shapes:

    try:
        file_stem = Path(file).stem
        gdf = gpd.read_file(file)

        gdf_olanda = gdf.to_crs(epsg=28992)
        gdf_olanda['area_mq'] = gdf_olanda.area
        gdf_olanda['area_km2'] = gdf_olanda['area_mq'] / 1_000_000

        temp_df = pd.DataFrame({
            'Name': file_stem,
            'Area_KM2': gdf_olanda['area_km2'],
            'Code' : gdf['OBJECTID'],
            'Type': gdf['GWS_GEWAS']
        })

        data_list.append(temp_df)
        #print(gdf_olanda['area_km2'])
    
    except Exception as e:
        print(f"Errore durante l'apertura dello shapefile {file}: {e}")

if data_list:
    df_finale = pd.concat(data_list, ignore_index=True)
    
    # Esporta in CSV
    df_finale.to_csv("areas_Flavoland.csv", index=False, sep=';', decimal='.')
    print("Processo completato! File salvato.")
else:
    print("Nessun dato elaborato.")
print()
```
